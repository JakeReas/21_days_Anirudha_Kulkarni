# Series Data Structure

The primary two components of pandas are the Series and DataFrame.

A **Series** is essentially a column, and a **DataFrame** is a multi-dimensional table made up of a collection of Series.
series-and-dataframe.width-1200.png

DataFrames and Series are quite similar in that many operations that you can do with one you can do with the other, such as filling in null values and calculating the mean.

A pandas Series is a one-dimensional labelled data structure which can hold data such as strings, integers and even other Python objects. It is built on top of numpy array and is the primary data structure to hold one-dimensional data in pandas.

In Python, a pandas Series can be created using the constructor pandas.Series().

**Syntax**

pandas.Series(data=None, index=None, dtype=None, name=None, copy=False)

**Parameters** 

**data**
 - array or dict or scalar value or iterables (default:None). It is used to populate the rows of the Series object.

**index**
 - array or index (default: None). It is used to label the rows of the Series. Its length must be the same as the object passed in the data parameter and all the values must be unique. Its default value is None but if the index parameter is not passed then a range of integer values from zero to one less than the total number of rows in the data object will be considered as the index.

**dtype**
 - String or numpy.dtype or ExtensionDtype (default: None). It is used to specify the data type of the Series which will be formed. If this parameter is not specified then the data type will be inferred from the values present in the series

**name**
 - String (default: None). It is used to give a name to the Series

**Copy - Boolean(default: False)**
- It is used to copy the input data

# What is Groupby in Pandas?

**GroupBy: Split, Apply, Combine**
Simple aggregations can give you a flavor of your dataset, but often we would prefer to aggregate conditionally on some label or index: this is implemented in the so-called groupby operation. 

Split, apply, combine
A canonical example of this split-apply-combine operation, where the "apply" is a summation aggregation, is illustrated in this figure:

03.08-split-apply-combine.png

This makes clear what the groupby accomplishes:

The split step involves breaking up and grouping a DataFrame depending on the value of the specified key.
The apply step involves computing some function, usually an aggregate, transformation, or filtering, within the individual groups.
The combine step merges the results of these operations into an output array.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(
    {"key1":list("aabbab"),
     "key2":["one","two","three"]*2,
     "data1":np.random.randn(6),
     "data2":np.random.randn(6)})
df
```

```python
group=df["data1"].groupby(df["key1"])
```

```python
group
```

```python
group.mean()
```

```python
ave=df["data1"].groupby([df["key1"],
                         df["key2"]]).mean()
ave
```

```python
ave.unstack()
```

```python
df.groupby("key1").mean()
```

```python
df.groupby(["key1","key2"]).mean()
```

# Pivot Tables

Pivot table in pandas is an excellent tool to summarize one or more numeric variable based on two other categorical variables. Pivot tables in pandas are popularly seen in MS Excel files. In python, Pivot tables of pandas dataframes can be created using the command: pandas.pivot_table.

You can aggregate a numeric column as a cross tabulation against two categorical columns.

**Syntax**

pandas.pivot_table(data, values=None, index=None, columns=None, aggfunc=mean, fill_value=None, margins=False, dropna=True, margins_name=All, observed=False)


```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(
    {"class":list("ABC")*4,
     "lesson":["math","stat"]*6,
     "sex":list("MFMM")*3,
     "sibling":[1,2,3]*4,
     "score":np.arange(40,100,5)})
```

```python
df
```

```python
df.groupby("lesson")["score"].mean()
```

```python
df.groupby(
    ["lesson",
     "class"])[
    "score"].aggregate("mean").unstack()
```

```python
df.pivot_table(
    "score",
    index="lesson",
    columns="class")
```

```python
df.pivot_table(
    ["sibling","score"],
    index=["class","lesson"],
    columns="sex")
```

```python
df.pivot_table(
    ["sibling","score"],
    index=["class","lesson"],
    columns="sex",margins=True)
```

```python
df.pivot_table(
    ["sibling","score"],
    index=["class","lesson"],
    columns="sex",fill_value=0)
```

## Multi-level pivot tables

```python
sib=pd.cut(df["sibling"],[0,2,3])
```

```python
df.pivot_table("score",
               ["lesson",sib],
               "class",fill_value=0)
```

```python
df.pivot_table(
    "score",
    index="lesson",
    columns="class")
```

```python
df.pivot_table(
    index="lesson",
    columns="class",
    aggfunc="sum")
```

```python
df.pivot_table(
    index="lesson",
    columns="class",
    aggfunc={"sibling":"max",
             "score":"sum"})
```

## Cross-Tabulations: Crosstab

# Time Series Basics with Pandas

## What is the time series?

A time series is any data set where the values are measured at different points in time. Many time series are uniformly spaced at a specific frequency, for example, hourly weather measurements, daily counts of web site visits, or monthly sales totals. Time series can also be irregularly spaced and sporadic, for example, timestamped data in a computer systems event log or a history of 911 emergency calls. Pandas time series tools apply equally well to either type of time series.

**How to simply create a time series?**

To show how to create a time series, first, lets import Pandas, Numpy, and datetime.

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

Lets create a date variable that contains the dates.

```python
date=[datetime(2020,1,5),
      datetime(2020,1,10),
      datetime(2020,1,15),
      datetime(2020,1,20),
      datetime(2020,1,25)] 
```

With this date object, let me create a series.

```python
ts=pd.Series(np.random.randn(5),index=date)
ts
```

Lets take a look at the index of this variable.

As you can see, the index data structure of this variable is a DatatimeIndex.

```python
ts.index 
```

## Time Series Data Structures

You can use three data structures to work with the time series. The first of these is Timestamps. Time stamps reference particular moments in time (e.g., June 5th, 2022 at 8:00 pm). This structure is similar to Pythons datetime module. But it is more useful. The index structure associated with this structure is DatetimeIndex.

For time Periods, Pandas provides the Period structure. The period consists of a time span with start and end dates. This structure is based on fixed interval dates. The index structure associated with this type is PeriodIndex.

For time deltas or durations, Pandas provides the Timedelta structure. Shows the time between two dates. The index structure associated with this structure is TimedeltaIndex.

The most used of these structures are Timestamps and DatetimeIndex objects. To convert a single date to Timestamps, you can use the to_datetime method.

```python
pd.to_datetime("01/01/2020") 
```

```python
dates=pd.to_datetime(
    [datetime(2020,7,5),
     "6th of July, 2020",
     "2020-Jul-7",
     "20200708"])
dates
```

```python
dates.to_period("D")
```

```python
dates-dates[0]
```

## Creating a Time Series

Pandas has various functions to create a date series. You can use the date_range method for timestamps, the period_range method for the period, and the timedelta_range method for time delta data.

The date_range method is used to get a fixed frequency DatetimeIndex. As you know, the range method in python and the arange method in numpy take the start point, end point, and step number parameters. Likewise, the date_range method takes the start date, the end date, and periods to create a date series. For example;

```python
pd.date_range("2020-08-15","2020-09-01") 
```

you can use the date range with the period parameter. To do this, you only need to enter the starting date.

```python
pd.date_range('2020-07-15', periods=10)
```

You can also use the freq parameter while creating a date range. By default, the frequency of the date_range method is daily. If you want to take the frequency as hours, you can use the H value that represents the hour.

```python
pd.date_range("2020-07-15",
              periods=10,
              freq="H")
```

To create a time series, you can also use the period_range and the timedelta_range methods.

```python
pd.period_range("2020-10", 
                periods=10,
                freq="M")
```

Lets create a time series with one hour increment

```python
pd.timedelta_range(0,periods=8,freq="H")
```

Time series can be indexed and selected like series structure in pandas. For example, lets assign index 1 of the ts time series to a variable.

```python
stamp=ts.index[1]
stamp
```

```python
# the value of this stamp
ts[stamp]
```

```python
# You can use a date index to select a value
ts["25.1.2020"]
```

```python
# OR
ts["20200125"]
```

```python
# long-time series
long_ts=pd.Series(
    np.random.randn(1000),
    index=pd.date_range("1/1/2020",
                        periods=1000))
long_ts.head()
```

```python
# Lets take a look at the values for 2020.
long_ts["2020"].head()
```

```python
# You can write this year with a month.
long_ts["2020-10"].head(15)
```

```python
# You can slice the time series.
long_ts[datetime(2022,9,20):] 
```

So far, you saw a time series that consists of a column. Now, Im going to want to show you how to work with the data frames. First, lets create a variable named date with the date_range method.

```python
date=pd.date_range("1/1/2020",
                   periods=100,
                   freq="W-SUN")
```

```python
# Lets create a data frame that indexes this date variable.
long_df=pd.DataFrame(np.random.randn(100,4),
                    index=date,
                    columns=list("ABCD"))
long_df.head()
```

```python
# look at the values in the 10th month of 2020 in this dataset.
long_df["2020-10"]
```

```python
# In some cases, the dates can be repeated
date=pd.DatetimeIndex(
    ["1/1/2020","1/2/2020","1/2/2020",
     "1/2/2020","1/3/2020"])
ts1=pd.Series(np.arange(5),index=date)
ts1
```

You can check whether there is a repeating index in the data or not with the is_unique attribute.

```python
ts1.index.is_unique 
```

False indicates that there are repeating indexes. Lets group repeated dates.

```python
group=ts1.groupby(level=0) 
```

Lets find out repeated dates with the count method.

```python
group.count()
```

Calculate means of the grouped dates.

```python
group.mean()
```
